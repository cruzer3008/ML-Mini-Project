```python
for i in range(len(df['ArrDelay'])):
    if((df['ArrDelay'].values)[i]>15.0):
        (df['ArrDelay'].values)[i] = 1.0
    else:
        (df['ArrDelay'].values)[i] = 0.0
```

```python
print(df['ArrDelay'].values)
```

The function written below shows the nuber of tuples which have target values 0 and 1 repectively

```python
count1 = 0
count0 = 0
for i in range(len(df['ArrDelay'])):
    if((df['ArrDelay'].values)[i] == 0.0):
        count0 = count0+1
    else:
        count1 = count1+1
print(count0,count1)
```

We can see that the dataset is **SKEWED** with a difference of 493266 tuples.

```python
1215012-721746
```

The method written down below will help remove the extra tuples with value 1.0, resulting in a balanced dataset

```python
indexes = []
for i in range(len(df['ArrDelay'].values)):
    if((df['ArrDelay'].values)[i] == 1.0):
        indexes.append(i)
```

```python
new_indexes = (indexes[:493266])
```

```python
df = df.drop(new_indexes)
```

```python
count1 = 0
count0 = 0
for i in range(len(df['ArrDelay'])):
    if((df['ArrDelay'].values)[i] == 0.0):
        count0 = count0+1
    else:
        count1 = count1+1
print(count0,count1)
```

```python
X = df[['Month', 'DayofMonth', 'DayOfWeek','FlightNum','CRSDepTime', 'DepDelay', 'TaxiOut', 'Distance','CRSArrTime',]]
y = df['ArrDelay']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
pred = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```
